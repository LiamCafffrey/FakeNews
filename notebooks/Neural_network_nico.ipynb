# Cleaning Data

```python
import string
import re

punc = string.punctuation + '“' + '”' + '’' + '‘'

def remove_punctuation(text):
    for punctuation in punc:
        text = text.replace(punctuation, '')
    return text

def lower_case(text):
    text = text.lower()
    return text

def remove_numbers(text):
    text = ''.join(word for word in text if not word.isdigit())
    return text


def rem_urls(text):
    return re.sub('https?:\S+','',text)

def rem_reuters(text):
    return re.sub('reuters','',text)
```

```python
data_working['title_text'] = data_working['title_text'].apply(remove_punctuation)
```

```python
data_working['title_text'] = data_working['title_text'].apply(lower_case)
```

```python
data_working['title_text'] = data_working['title_text'].apply(remove_numbers)
```

```python
data_working['title_text'] = data_working['title_text'].apply(rem_reuters)
```

```python
data_working['title_text'] = data_working['title_text'].apply(rem_urls)
```

# Removing stopwords and Wordlemmatizer

```python
import nltk
from nltk.corpus import stopwords

stop = set(stopwords.words('english'))

def rem_stopwords(df_news):
    
    words = [ch for ch in df_news if ch not in stop]
    words= "".join(words).split()
    words= [words.lower() for words in df_news.split()]
    
    return words 
```

```python
data_working['title_text'].apply(rem_stopwords)
```

```python
data_working.score.value_counts()
```

```python
from nltk.stem import WordNetLemmatizer
#nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()

def lemmatize_words(text):
    lemmas= []
    for word in text.split():
        lemmas.append(lemmatizer.lemmatize(word))
    return " ".join(lemmas)
```
