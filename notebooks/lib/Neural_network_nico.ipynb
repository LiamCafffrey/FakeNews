# Cleaning Data

```python
import string
import re

punc = string.punctuation + '“' + '”' + '’' + '‘'

def remove_punctuation(text):
    for punctuation in punc:
        text = text.replace(punctuation, '')
    return text

def lower_case(text):
    text = text.lower()
    return text

def remove_numbers(text):
    text = ''.join(word for word in text if not word.isdigit())
    return text


def rem_urls(text):
    return re.sub('https?:\S+','',text)

def rem_reuters(text):
    return re.sub('reuters','',text)
```

```python
data['title_text'] = data['title_text'].apply(remove_punctuation)
```

```python
data['title_text'] = data['title_text'].apply(lower_case)
```

```python
data['title_text'] = data['title_text'].apply(remove_numbers)
```

```python
data['title_text'] = data['title_text'].apply(rem_reuters)
```

```python
data['title_text'] = data['title_text'].apply(rem_urls)
```

```python
data.head()
```

# Removing stopwords and Wordlemmatizer

```python
import nltk
from nltk.corpus import stopwords

stop = set(stopwords.words('english'))

def rem_stopwords(df_news):
    
    words = [ch for ch in df_news if ch not in stop]
    words= "".join(words).split()
    words= [words.lower() for words in df_news.split()]
    
    return words 
```

```python
data['title_text'].apply(rem_stopwords)
```

```python
from nltk.stem import WordNetLemmatizer
#nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()

def lemmatize_words(text):
    lemmas= []
    for word in text.split():
        lemmas.append(lemmatizer.lemmatize(word))
    return " ".join(lemmas)
```

```python
data['title_text'].apply(lemmatize_words)
```

```python
x = data['title_text']
y = data['score']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=0.3)
```

```python
from keras.preprocessing.text import Tokenizer

max_vocab = 30000
tokenizer = Tokenizer(num_words=max_vocab)
tokenizer.fit_on_texts(x_train)
```

```python
# tokenize the text into vectors 
X_train = tokenizer.texts_to_sequences(x_train)
X_test = tokenizer.texts_to_sequences(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np

X_train = tf.keras.preprocessing.sequence.pad_sequences(X_train, padding='post', maxlen=256)
X_test = tf.keras.preprocessing.sequence.pad_sequences(X_test, padding='post', maxlen=256)
```

# Building the model

```python
from tensorflow.keras import Sequential
from tensorflow.keras import layers


import tensorflow as tf




model = Sequential()
model.add(layers.Embedding(max_vocab, output_dim=128))  # We have 2000 words in the vocabulary, 
model.add(layers.Masking())                                                     # and each word is represnted by a vector of size 30
model.add(layers.LSTM(128, activation='tanh'))
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping

es = EarlyStopping(patience=5, restore_best_weights=True)

model.fit(X_train, y_train, 
          epochs=5, batch_size=16, 
          validation_split=0.3,
          callbacks=[es])
```
